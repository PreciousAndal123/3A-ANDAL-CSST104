1. Importing Libraries

```python
import pandas as pd #Data Manipulation
import matplotlib.pyplot as plt #Data Visualization
```

2. Loading Datasets

```python
#import sales_data_2.csv
data = pd.read_csv('sales_data_2.csv')
```

3. Data Loading and Data Analysis

```python
statistics = data.describe()
```

4. Data Manipulation

```python
product_revenue = data.groupby('Product')['Revenue'].sum()

highest_revenue_product = product_revenue.idxmax()

average_quantity_per_day = data['Quantity Sold'].mean()
```

5. Data Visualization

```python
product_revenue.plot(kind='bar')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.title('Revenue by Product')
plt.xticks(rotation=50)
plt.tight_layout()
plt.show()
```

6. Analysis Report

```python
report = """
Analysis Report

1. Data Loading and Initial analysis:
    Basic statistics of the Datasets:
    {}
2. Datav Manipulation:
    Total Revenue for each product:
    {}
    Product with the highest revenue: {}
    Average quantity sold per day: {:.2f}
3. Data Visualization:
    Bar Chart Displaying revenue for each product is attached.

4. Insights:
   -The product {} generates the highest revenue.
   -On Average, {:.2f} units sold per day.

""".format(statistics, product_revenue, highest_revenue_product, average_quantity_per_day,
           highest_revenue_product, average_quantity_per_day)
print(report)
```

7. Saving the Report to a file

```python
with open('report.txt','w') as f:
  f.write(report)
```
